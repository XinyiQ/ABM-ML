```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\ITC\ABM\CASE STUDY_\TRY\20144444444.csv',header=None)
array = df.to_numpy()
array[np.isnan(array)] = 0 # There are some null values, they are replaced to 0
```

```python
labels = ['Tike_bites_count', 'Landuse', 'Agent_count', 'Tick_abundance', 'Working', 'Gardening', 'Picnic', 'Hiking', 'Playing', 'Sunbathing']
plt.suptitle("Histogram of the weather variables")
plt.figure(figsize=(15,15))
plt.subplots_adjust(hspace=0.5, wspace=0.5)
for i in range(1, 11): 
    # Store in a variable called “curr” the i-th colum 
    curr = array[:,i-1:i]
    plt.subplot(3, 4, i)
    # Call the function plt.hist() with “curr” and 25 bins
    plt.hist(curr,bins=25)
    # Add proper labels to the subplot
    plt.xlabel("Variables")
    plt.ylabel("Frequency")
    # Add grid lines to the subplot 
    plt.grid()
    # Call the plt.title() function to add the (i – 1) label 
    plt.title(str(labels[i-1]))
    # Show the chart
plt.show()
```
